```python
from bs4 import BeautifulSoup
```

```python
#load response object into soup
soup = BeautifulSoup(djiResponse.text)
print(soup.prettify())
```

```python
# parse text containing tickers
import re
row1 = soup.find_all(href=re.compile("https://www.nyse.com/quote/XNYS:"))
row2 = soup.find_all(href=re.compile("http://www.nasdaq.com/symbol/"))
row = row1+row2
print(row)
```

```python
# extract text between tags for tickers
dji_tickers = []
for t in row:
    if t.getText() == None:
        pass
    else:
        dji_tickers.append(t.getText())
print(dji_tickers)
```

## Part 2a. Grabbing the Data 

Using the data structure of the list of stocks from above.

```python
#import library for timeseries module 

from alpha_vantage.timeseries import TimeSeries

ts = TimeSeries(key='GNAZYAP4YGZWRTPF', output_format='pandas')
```

```python
#Randomly select 3 stocks from the list
import random
data_request = random.sample(dji_tickers, 3)
print(data_request)
```

```python
#get daily price data for stocks in the date range and write to .csv file
import time

for dji_tickers in data_request:
    #Prints every 5 seconds
    data, meta_data = ts.get_daily(symbol= dji_tickers , outputsize='full')
    data_daily = data[(data.index >= '2014-01-01') & (data.index <= '2019-12-31')]
    #Save file to csv
    filename = dji_tickers +'_daily.csv'
    data_daily.to_csv(filename)
    #API limits to 5 calls per minute
    time.sleep(20)
```

```python
#Load up the .csv files that were created from the list of stocks
import pandas as pd
import os

file_list_1=os.listdir()
file_list_2=[]

for filename in file_list_1:
    if '_daily.csv' in filename:
        file_list_2.append(filename)

#read file from file list and convert to daily price dataframe
for filename in file_list_2:
    data=pd.read_csv(filename)
    
    data.index = pd.to_datetime(data['date'])
    data = data.drop(columns=['date'])
    print('---------------------------'+filename.strip('_daily.csv')+'---------------------------')
    print(data)
    print('\n')
```

Calculate the following from the data.

Print the following:

```python
import numpy as np
```

```python
#get daily adjusted price data for stocks in the date range and write to .csv file
for dji_tickers in data_request:
    data_adjusted_daily, meta_data = ts.get_daily_adjusted(symbol= dji_tickers , outputsize='full')
    data_adjusted_daily = data_adjusted_daily[(data_adjusted_daily.index >= '2014-01-01') & (data_adjusted_daily.index <= '2019-12-31')]
    #Save file to csv
    filename_adjusted= dji_tickers +'_adjusted.csv'
    data_adjusted_daily.to_csv(filename_adjusted)
    #API limits to 5 calls per minute
    time.sleep(20)
```

```python
#Load up the files that were created from the list of stocks
import pandas as pd
import os

file_list_1=os.listdir()
file_list_2=[]

for filename in file_list_1:
    if '_adjusted.csv' in filename:
        file_list_2.append(filename)

#initialize resultant dataframe for adjusted close of 3 stocks
result_df = pd.DataFrame()

#create resultant dataframe for log return of adjusted close of stocks
for filename in file_list_2:
    adjusted_df=pd.read_csv(filename)
    adjusted_df.index = pd.to_datetime(adjusted_df['date'])
    
    adjusted_df["log_adj_df"] = adjusted_df['5. adjusted close'].apply(lambda x: np.log(x)) - adjusted_df['5. adjusted close'].shift(1).apply(lambda x: np.log(x))
    result_df[filename.strip('_adjusted.csv')] = adjusted_df["log_adj_df"]
```

```python
#Log Returns of Adjusted Price Data
print('--------Log Returns of Adjusted Price---------')
print(result_df)
```

```python
#Average of Log Returns
print("---------------Mean---------------")
print (result_df.mean())
```

```python
#Standard Deviation of Log Returns
print('---------------Std Dev---------------')
print (result_df.std())
```

```python
#Correlation Matrix of Log Returns
print("---------------Correlation Matrix---------------")
print(result_df.corr())
```

```python
#Load up the files that were created from the list of stocks

import pandas as pd
import os

file_list_1=os.listdir()
file_list_2=[]

for filename in file_list_1:
    if '_adjusted.csv' in filename:
        file_list_2.append(filename)

#initialize resultant dataframe for adjusted close of 3 stocks
result_df = pd.DataFrame()

#create resultant dataframe for log return of adjusted close of stocks
for filename in file_list_2:
    adjusted_df=pd.read_csv(filename)
    adjusted_df.index = pd.to_datetime(adjusted_df['date'])
    adjusted_df["log_adj_df"] = adjusted_df['5. adjusted close'].apply(lambda x: np.log(x)) - adjusted_df['5. adjusted close'].shift(1).apply(lambda x: np.log(x))
    result_df[filename.strip('_adjusted.csv')] = adjusted_df["log_adj_df"]

#randomly select one stock column for plotting
result_df_1 = result_df.sample(axis=1)
result_df_1.dropna(inplace=True)
result_df_1_columns = result_df_1.columns
```

```python
#Plot Histogram of the log returns
plt.hist(result_df_1[result_df_1_columns[0]], bins=30, facecolor="blue", edgecolor="white", alpha=0.7)
plt.xlabel('Log return')
plt.ylabel('Frequency')
plt.title(f"Log Returns of Stock: {result_df_1_columns[0]}")
```
